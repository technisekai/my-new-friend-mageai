```python
import requests

api_url = 'https://openlibrary.org/subjects/science.json'
response = requests.get(api_url)
book_data = response.json() 

book_data_works = book_data["works"]
book_data_works[0]["new"] = 12345
```

```python
from data_warehouse.cores.postgresql import postgresql_connection, insert_into_postgres, create_table_postgresql, add_columns_postgresql
from psycopg2 import errors

dwh_connect = postgresql_connection(
    host='localhost',
    port=5432,
    user='admin',
    password='admin123',
    database='public'
)

create_table_postgresql(
    book_data_works[0],
    'public',
    'test_book_works',
    dwh_connect
)

for data in book_data_works:
    print(data)
    add_columns_postgresql(
        source_data=data,
        destination_table_name='test_book_works',
        destination_schema_name='public',
        destination_connect=dwh_connect
    )

insert_into_postgres(
    destination_connect=dwh_connect,
    destination_schema_name='public',
    destination_table_name='test_book_works',
    data=book_data_works,
    mode='single'
)

dwh_connect.close()
```
